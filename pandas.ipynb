### 1. Compute the total number of unique apps from the dataset


```python
import csv
import numpy as np
with open(r'C:\Users\Archana\Desktop\py1\Applestore.csv',encoding='utf8') as csvfile:
    csv_reader = list(csv.reader(csvfile))
    
    unique_list=[]
    
    x=np.array(csv_reader)
    
    for i in range(1,len(x)):
        
        if x[i][1] not in unique_list:
            unique_list.append(x[i][1])
    print(len(unique_list))
        
```

### 3.Categorise the dataset based on content rating into the following
- Number of apps with content rating 4+
- Number of apps with content rating 9+
- Number of apps with content rating 12+
- Number of apps with content rating 17+

```python
import csv
import numpy as np
with open(r'C:\Users\Archana\Desktop\py1\Applestore.csv',encoding='utf8') as csvfile:
    csv_reader = list(csv.reader(csvfile))
    unique1=[]
    unique2=[]
    unique3=[]
    unique4=[]
    x=np.array(csv_reader)
    for i in range(1,len(x)):
        if '4+' in x[i][10]:
            unique1.append('4+') 
        elif '9+' in x[i][10]:
            unique2.append('9+') 
        elif '12+' in x[i][10]:
            unique3.append('12+') 
            
        else:
            unique4.append('17+')
    print("4+ = ",len(unique1))
    print("9+ = ",len(unique2))
    print("12+ = ",len(unique3))
    print("17+ = ",len(unique4))
```
